## marker

```python
y=np.array([2,3,6,9,11])
plt.plot(y,marker='o',color='orange') ##if we use marker it will get patter like this
plt.show()## if we write only o it just plot dot but marker plot with dot and line combination
```

### set font properties title and label

```python
x=np.array([2000,2002,2004,2006,2010])
y=np.array([240,250,260,270,280])

font1={'family':"arial",'color':'blue',"size":20}
font2={'family':"times new roman",'color':'red',"size":15}

plt.title("sales data",fontdict=font1)
plt.xlabel("year",fontdict=font2)
plt.ylabel("sales progress",fontdict=font2)

plt.plot(x,y)
plt.show()
```

```python
x=np.array([2000,2002,2004,2006,2010])
y=np.array([240,250,260,270,280])

plt.title("sales data")
plt.xlabel("year")
plt.ylabel("sales progress")

plt.plot(x,y)
plt.show()
```

### position the title

```python
x=np.array([2000,2002,2004,2006,2010])
y=np.array([240,250,260,270,280])

font1={'family':"arial",'color':'blue',"size":20}
font2={'family':"times new roman",'color':'red',"size":15}

plt.title("sales data",fontdict=font1,loc='left')
plt.xlabel("year",fontdict=font2)
plt.ylabel("sales progress",fontdict=font2)

plt.plot(x,y)
plt.show()
```

## bar graph

```python
#A bar graph uses bars to compare data among different categories.it is well suited when you want to measure the changes overa period of time.it can be represented horizontally or vertically.
#The important thing to keep in mind is that longer the bar ,greater the value.
#The bar() function accepts three arguments -categorical,variables,values and color.
#The barh() function help us to create horizontal bar graph.
#The bar() takes the keyword argument width to set the width of the bars.
#The barh() takes the keyword argument height to set the height of the bars.
#The default height value is 0.8
```

```python
plt.bar(['airtel',"jio","vodafone","idea","bsnl"],[45,55,75,65,85],label="isp",color="r",width=0.5)##use this graph one parameter one value for each
plt.legend()
plt.xlabel("days")
plt.ylabel("usage per day")
plt.title("data usage")
plt.show()
```

### stacked vertical bar

```python
plt.style.use("seaborn-darkgrid")
x=["airtel","jio","vodafone","idea"]
y=[17,18,29,40]
y1=[20,21,22,23]
plt.figure(figsize=(5,7))
plt.bar(x,y,label="2019",width=0.5,color="r")
plt.bar(x,y1,label="2020",width=0.5,bottom=y,color="b")
plt.xlabel("isp")
plt.ylabel("sales")
plt.title("sales by year")
plt.legend()
plt.show()
```

```python
plt.style.use("seaborn-darkgrid")
x=["airtel","jio","vodafone","idea"]
y=np.array([17,18,29,40])
y1=np.array([20,21,22,23])
y2=np.array([4,5,6,25])
plt.figure(figsize=(5,7))
plt.bar(x,y,label="2019",width=0.5,color="r")
plt.bar(x,y1,label="2020",width=0.5,bottom=y,color="b")
plt.bar(x,y2,label="2021",width=0.5,bottom=y+y1,color="g")##bottom for printing one graph on bottom of other and soon
plt.xlabel("isp")
plt.ylabel("sales")
plt.title("sales by year")
plt.legend()
plt.show()
```

```python
plt.style.use("seaborn-darkgrid")
x=["airtel","jio","vodafone","idea"]
y=[17,18,29,40]
y1=[20,21,22,23]
plt.figure(figsize=(5,7))
plt.barh(x,y,label="2019",color="r")
plt.barh(x,y1,label="2020",left=y,color="b")##using left so figure like this
plt.xlabel("isp")
plt.ylabel("sales")
plt.title("sales by year")
plt.legend()
plt.show()
```

```python
plt.style.use("seaborn-darkgrid")
x=["airtel","jio","vodafone","idea"]
y=np.array([17,18,29,40])
y1=np.array([20,21,22,23])
y2=np.array([4,5,6,25])
plt.figure(figsize=(5,7))
plt.barh(x,y,label="2019",color="r")
plt.barh(x,y1,label="2020",left=y,color="b")
plt.barh(x,y2,label="2021",left=y+y1,color="g")##bottom for printing one graph on bottom of other and soon
plt.xlabel("isp")
plt.ylabel("sales")
plt.title("sales by year")
plt.legend()
plt.show()
```
